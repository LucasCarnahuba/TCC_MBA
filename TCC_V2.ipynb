```python
import requests
import webbrowser
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN, LSTM, GRU
from tensorflow.keras.optimizers import Adam
from sklearn.preprocessing import MinMaxScaler
import tkinter as tk
from tkinter import ttk, simpledialog, messagebox
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

# URL do site
url = "https://statusinvest.com.br/fundos-de-investimento"

# Cabeçalhos para simular um navegador real
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36',
    'Accept-Language': 'en-US,en;q=0.9',
    'Accept-Encoding': 'gzip, deflate, br',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
    'Connection': 'keep-alive',
}

# Fazer a requisição
response = requests.get(url, headers=headers)

# Verificar se a requisição foi bem-sucedida
if response.status_code == 200:
    # Parsear o HTML da página
    soup = BeautifulSoup(response.content, 'html.parser')
    print("Página carregada com sucesso!")
else:
    print(f"Erro ao acessar a página. Código HTTP: {response.status_code}")

# Global variables for results and data
results = {}
categorias = {}
historico_fundos = pd.DataFrame(columns=['Razão Social', 'CNPJ'])
data = pd.DataFrame(columns=['Razão Social', 'DT_COMPTC', 'Rentabilidade_Acumulada'])

# Função para pegar CNPJ e Razão Social de um fundo
def obter_razao_cnpj_fundo(fundo_url, headers):
    response = requests.get(fundo_url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Encontrar a razão social (tag <span> dentro do <h4>)
    razao_social_tag = soup.find('h4', {'class': 'mt-0 mb-0 fs-4 lh-4 fs-md-5 lh-md-5'})
    cnpj_tag = soup.find('h5', {'class': 'fw-600'})  # CNPJ geralmente é o primeiro <h5> com essa classe após "CNPJ"

    # Extrair os textos das tags, se encontradas
    if razao_social_tag and cnpj_tag:
        razao_social = razao_social_tag.find('span', {'class': 'd-block fw-600 text-main-green-dark'}).text.strip() if razao_social_tag.find('span', {'class': 'd-block fw-600 text-main-green-dark'}) else None
        cnpj = cnpj_tag.text.strip() if cnpj_tag else None
        return {
            'Razão Social': razao_social,
            'CNPJ': cnpj
        }
    return None

# Função para carregar categorias disponíveis
def carregar_categorias(soup):
    global categorias
    categorias = {i: h2.text.strip() for i, h2 in enumerate(soup.find_all('h2'), start=1)}
    return categorias

# Parte 1: Seleção de Categorias
def selecionar_categoria():
    carregar_categorias(soup)
    categoria_window = tk.Toplevel(root)
    categoria_window.title("Seleção de Categoria")
    tk.Label(categoria_window, text="Categorias disponíveis:").pack()

    for idx, nome in categorias.items():
        btn = ttk.Button(categoria_window, text=f"{idx}: {nome}", command=lambda i=idx: confirmar_categoria(i))
        btn.pack(fill="x", padx=10, pady=5)

# Função para exibir os fundos e adicionar o botão para gerar previsões
def exibir_fundos(categoria_escolhida, fundos_list):
    fundos_window = tk.Toplevel(root)
    fundos_window.title(f"Fundos de {categoria_escolhida}")

    # Exibe o título da categoria
    tk.Label(fundos_window, text=f"Fundos da categoria: {categoria_escolhida}").pack(pady=10)

    # Exibe informações dos fundos
    for idx, fundo in enumerate(fundos_list, start=1):
        fundo_info = f"{idx}. Nome: {fundo['Nome']} | Patrimônio: {fundo['Patrimônio']} | Rentabilidade: {fundo['Rentabilidade']}"
        tk.Label(fundos_window, text=fundo_info).pack(pady=5)

        # Botão para abrir o link do fundo no navegador
        info_button = ttk.Button(fundos_window, text="Ver detalhes", command=lambda url=fundo['Link']: webbrowser.open(url))
        info_button.pack(pady=5)
        
    # Botão para abrir a nova tela de previsões e projeções
    btn_previsoes = ttk.Button(fundos_window, text="Abrir Tela de Previsões e Projeções", command=abrir_tela_previsoes)
    btn_previsoes.pack(pady=20)
    
    # Gera automaticamente o gráfico de rentabilidade acumulada
    rentabilidade_acumulada()

    
    
def confirmar_categoria(escolha):
    global categorias
    categoria_escolhida = categorias.get(escolha)
    if not categoria_escolhida:
        messagebox.showerror("Erro", "Categoria inválida.")
    else:
        messagebox.showinfo("Categoria Selecionada", f"Você escolheu: {categoria_escolhida}")
        # Localiza a seção correspondente aos fundos da categoria escolhida
        section = soup.find('h2', text=categoria_escolhida).find_next('div', class_='list')

        if section:
            # Lista para armazenar os fundos
            fundos_list = []
            fundos_links = []  # Lista para armazenar os links
            # Captura os fundos na seção
            fundos = section.find_all('a', class_='d-flex align-items-center waves-effect waves-on-white-bg')

            # Pega os top 10 (ou menos, caso não existam 10 fundos)
            for fundo in fundos[:10]:
                # Nome do fundo
                nome = fundo.find('h4').text.strip()

                # Patrimônio
                patrimonio = fundo.find('span', class_='d-flex fw-700 other-value')
                patrimonio = patrimonio.text.strip() if patrimonio else "Não informado"
                patrimonio = "R$ " + patrimonio.replace('\n', '').replace(' ', '').replace('M', ' M').replace('B', ' B')
                # Rentabilidade
                rentabilidade_span = fundo.find('span', class_='value align-items-center d-flex')
                rentabilidade = (
                    rentabilidade_span.text.strip()
                    .replace('\n', '')
                    .replace('arrow_upward', '')
                    .strip()
                    if rentabilidade_span
                    else "Não informado"
                )

                # Pega o link do fundo
                link_fundo = fundo.get('href')
                if link_fundo:
                    fundo_url = "https://statusinvest.com.br" + link_fundo
                    razao_cnpj = obter_razao_cnpj_fundo(fundo_url, headers)
                    fundos_list.append({
                        'Nome': nome,
                        'Patrimônio': patrimonio,
                        'Rentabilidade': rentabilidade,
                        'Razão Social': razao_cnpj['Razão Social'] if razao_cnpj else 'Não encontrado',
                        'CNPJ': razao_cnpj['CNPJ'] if razao_cnpj else 'Não encontrado',
                        'Link': fundo_url
                    })

            # Atualiza o DataFrame global com os CNPJs dos fundos
            global historico_fundos
            historico_fundos = pd.DataFrame(fundos_list)

            # Exibe os fundos na janela de Tkinter
            exibir_fundos(categoria_escolhida, fundos_list)

        else:
            print("Seção correspondente à categoria escolhida não encontrada.")

def rentabilidade_acumulada():
    # Verifica se o DataFrame de fundos foi preenchido
    if historico_fundos.empty:
        messagebox.showerror("Erro", "Nenhuma categoria foi selecionada ou nenhum fundo foi carregado.")
        return

    # Filtra apenas os CNPJs do histórico_fundos para otimizar a leitura
    cnpjs_historico = historico_fundos['CNPJ'].unique()

    # Lista para armazenar os pedaços do DataFrame
    tabela_completa = []

    # Define o intervalo de ANOMES que você quer carregar
    for ano_mes in range(202301, 202412):  # De 202301 até 202411
        ano, mes = divmod(ano_mes, 100)

        # Verifica se o mês é válido (de 01 a 12)
        if mes < 1 or mes > 12:
            continue  # Pula se o mês não for válido

        # Constrói o nome do arquivo
        arquivo_csv = f'INFORME_DIARIO/inf_diario_fi_{ano_mes}.csv'

        try:
            # Usando chunksize para carregar o arquivo em partes
            chunk_iter = pd.read_csv(arquivo_csv, sep=';', encoding='ISO-8859-1', 
                                     chunksize=100000)  # Ajuste o chunksize conforme necessário

            for chunk in chunk_iter:
                # Verifica a partir de qual coluna filtrar (antes ou depois de 202404)
                if ano_mes < 202404:
                    chunk_filtrado = chunk[chunk['CNPJ_FUNDO'].isin(cnpjs_historico)]
                    join_column = 'CNPJ_FUNDO'
                else:
                    chunk_filtrado = chunk[chunk['CNPJ_FUNDO_CLASSE'].isin(cnpjs_historico)]
                    join_column = 'CNPJ_FUNDO_CLASSE'

                # Realiza o inner join com o histórico_fundos
                df_merged = pd.merge(chunk_filtrado, historico_fundos, 
                                      left_on=join_column, right_on='CNPJ', how='inner')

                # Remove a coluna CNPJ após o merge
                df_merged = df_merged.drop(columns=['CNPJ'])

                # Adiciona o pedaço do DataFrame ao resultado final
                tabela_completa.append(df_merged)

        except FileNotFoundError:
            print(f"Arquivo não encontrado: {arquivo_csv}")
            continue  # Pula o arquivo que não foi encontrado

    df_final = pd.concat(tabela_completa, ignore_index = True)

    # Preenchendo coluna CNPJ_FUNDO
    df_final['CNPJ_FUNDO'] = df_final['CNPJ_FUNDO'].combine_first(df_final['CNPJ_FUNDO_CLASSE'])

    # Ordenar o DataFrame por CNPJ_FUNDO e DT_COMPTC (data) para garantir que os dados estão na ordem correta
    df_final['DT_COMPTC'] = pd.to_datetime(df_final['DT_COMPTC'])
    df_final = df_final.sort_values(by=['CNPJ_FUNDO', 'DT_COMPTC'])

    # Inicializando a coluna de rentabilidade acumulada
    df_final['Rentabilidade_Acumulada'] = 0.0

    # Loop para calcular a rentabilidade diária acumulada por fundo
    for cnpj in df_final['CNPJ_FUNDO'].unique():
        # Filtra o DataFrame para cada fundo
        fundo = df_final[df_final['CNPJ_FUNDO'] == cnpj]

        # Loop para calcular a rentabilidade diária
        for i in range(1, len(fundo)):
            # Rentabilidade diária
            rentabilidade_diaria = (fundo.iloc[i]['VL_QUOTA'] - fundo.iloc[i-1]['VL_QUOTA']) / fundo.iloc[i-1]['VL_QUOTA']

            # Rentabilidade acumulada
            fundo.loc[fundo.index[i], 'Rentabilidade_Acumulada'] = fundo.iloc[i-1]['Rentabilidade_Acumulada'] + rentabilidade_diaria

        # Atualiza o DataFrame original com as rentabilidades acumuladas calculadas
        df_final.loc[df_final['CNPJ_FUNDO'] == cnpj, 'Rentabilidade_Acumulada'] = fundo['Rentabilidade_Acumulada']
    
    global data
    data = df_final[['Razão Social', 'DT_COMPTC', 'Rentabilidade_Acumulada']]
    
    # Plotando a rentabilidade acumulada para cada fundo
    fig, ax = plt.subplots(figsize=(10, 6))

    for cnpj, fundo in df_final.groupby('CNPJ_FUNDO'):
        # Ajustar a razão social para abreviar se for muito longa
        razao_social_abrev = fundo['Razão Social'].iloc[0][:20] + '...' if len(fundo['Razão Social'].iloc[0]) > 20 else fundo['Razão Social'].iloc[0]
        ax.plot(fundo['DT_COMPTC'], fundo['Rentabilidade_Acumulada'], label=razao_social_abrev)

    ax.set_title('Rentabilidade Acumulada dos Fundos ao Longo do Tempo')
    ax.set_xlabel('Data de Competência')
    ax.set_ylabel('Rentabilidade Acumulada')

    # Ajustando o tamanho da fonte da legenda
    ax.legend(title="Razão Social", bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)

    # Garantir que o layout não será cortado
    fig.tight_layout()

    # Criando uma nova janela para exibir o gráfico
    graph_window = tk.Toplevel(root)
    graph_window.title("Rentabilidade Acumulada dos Fundos")

    # Exibindo o gráfico na nova janela
    canvas = FigureCanvasTkAgg(fig, master=graph_window)  # Embed the figure in the Toplevel window
    canvas.draw()

    # A nova janela vai expandir para o tamanho do gráfico
    canvas.get_tk_widget().pack(fill='both', expand=True)

    # Ajusta o layout da janela para evitar cortes
    graph_window.update_idletasks()
    graph_window.geometry(f"{graph_window.winfo_width()}x{graph_window.winfo_height()}")
    
# Função para preparar os dados
def prepare_data(series, n_steps):
    X, y = [], []
    for i in range(len(series) - n_steps):
        X.append(series[i:i + n_steps])
        y.append(series[i + n_steps])
    return np.array(X), np.array(y)



def create_model(model_type):
    n_steps = 30
    model = Sequential()
    if model_type == 'RNN':
        model.add(SimpleRNN(50, activation='relu', input_shape=(n_steps, 1)))
    elif model_type == 'LSTM':
        model.add(LSTM(50, activation='relu', input_shape=(n_steps, 1)))
    elif model_type == 'GRU':
        model.add(GRU(50, activation='relu', input_shape=(n_steps, 1)))
    model.add(Dense(1))
    model.compile(optimizer=Adam(learning_rate=0.001), loss='mse')
    return model

# Função de previsões com o gráfico
def previsoes():
    n_steps = 30  # Janela de tempo

    # Criando uma nova janela para exibir os gráficos com a barra de rolagem
    previsoes_window = tk.Toplevel(root)
    previsoes_window.title("Modelos para os Fundos")

    # Canvas com barra de rolagem
    canvas_frame = tk.Frame(previsoes_window)
    canvas_frame.pack(fill=tk.BOTH, expand=True)

    canvas = tk.Canvas(canvas_frame)
    scrollbar = tk.Scrollbar(canvas_frame, orient="vertical", command=canvas.yview)
    canvas.configure(yscrollcommand=scrollbar.set)
    scrollbar.pack(side=tk.RIGHT, fill=tk.Y)
    canvas.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)

    # Frame para colocar os widgets no canvas
    frame = tk.Frame(canvas)
    canvas.create_window((0, 0), window=frame, anchor="nw")

    for fundo in data['Razão Social'].unique():
        print(f'Treinando modelos para o fundo: {fundo}')
        fundo_data = data[data['Razão Social'] == fundo].sort_values(by='DT_COMPTC')
        series = fundo_data['Rentabilidade_Acumulada'].values

        scaler = MinMaxScaler()
        series_scaled = scaler.fit_transform(series.reshape(-1, 1)).flatten()

        X, y = prepare_data(series_scaled, n_steps)
        split = int(0.8 * len(X))
        X_train, X_test = X[:split], X[split:]
        y_train, y_test = y[:split], y[split:]

        X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
        X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

        models = {}
        predictions = {}
        metrics = {}

        for model_type in ['RNN', 'LSTM', 'GRU']:
            model = create_model(model_type)
            model.fit(X_train, y_train, epochs=20, verbose=0, batch_size=32)
            y_pred = model.predict(X_test).flatten()

            y_pred_rescaled = scaler.inverse_transform(y_pred.reshape(-1, 1)).flatten()
            y_test_rescaled = scaler.inverse_transform(y_test.reshape(-1, 1)).flatten()

            mae = mean_absolute_error(y_test_rescaled, y_pred_rescaled)
            mse = mean_squared_error(y_test_rescaled, y_pred_rescaled)

            models[model_type] = model
            predictions[model_type] = y_pred_rescaled
            metrics[model_type] = {'MAE': mae, 'MSE': mse}

        best_model_type = min(metrics, key=lambda x: metrics[x]['MAE'])
        global results
        results[fundo] = {
            'best_model': models[best_model_type],
            'best_model_type': best_model_type,
            'metrics': metrics,
            'data': fundo_data,
            'scaler': scaler,
        }

        # Gerando o gráfico para o fundo
        fig, ax = plt.subplots(figsize=(12, 6))
        ax.plot(fundo_data['DT_COMPTC'].iloc[-len(y_test):], y_test_rescaled, label='Real')
        for model_type in predictions.keys():
            ax.plot(fundo_data['DT_COMPTC'].iloc[-len(y_test):], predictions[model_type], label=model_type)

        ax.set_title(f'Previsões para o fundo {fundo}')
        ax.set_xlabel('Data')
        ax.set_ylabel('Rentabilidade Acumulada')
        ax.legend()

        # Integrando o gráfico com o Tkinter
        canvas_fig = FigureCanvasTkAgg(fig, master=frame)
        canvas_fig.get_tk_widget().pack(pady=10)
        canvas_fig.draw()

        # Adicionando tabela com MAE e MSE compacta
        tree = ttk.Treeview(frame, columns=("Modelo", "MAE", "MSE"), show="headings", height=3)
        tree.heading("Modelo", text="Modelo")
        tree.heading("MAE", text="MAE")
        tree.heading("MSE", text="MSE")

        # Formatar colunas para ficarem menores
        tree.column("Modelo", width=100, anchor="center")
        tree.column("MAE", width=80, anchor="center")
        tree.column("MSE", width=80, anchor="center")

        # Encontrar os menores valores de MAE e MSE
        min_mae = min(metrics[model]['MAE'] for model in metrics)
        min_mse = min(metrics[model]['MSE'] for model in metrics)

        # Inserir as métricas na tabela com destaque
        for model_type in metrics.keys():
            mae = metrics[model_type]['MAE']
            mse = metrics[model_type]['MSE']

            # Destacar menores valores de MAE e MSE
            tags = []
            if mae == min_mae:
                tags.append("min_mae")
            if mse == min_mse:
                tags.append("min_mse")

            # Inserir valores na tabela com tags para destaque
            tree.insert("", tk.END, values=(model_type, f"{mae:.4f}", f"{mse:.4f}"), tags=tags)

        # Adicionar estilos para destaque
        style = ttk.Style()
        style.configure("Treeview.Heading", font=("Arial", 10, "bold"))
        tree.tag_configure("min_mae", background="lightgreen")  # Destacar menor MAE
        tree.tag_configure("min_mse", background="lightblue")   # Destacar menor MSE

        tree.pack(pady=10)

    # Atualizando a região visível do canvas para ajustar a rolagem
    frame.update_idletasks()
    canvas.config(scrollregion=canvas.bbox("all"))
    
# Função para realizar projeção e exibir gráficos em uma nova tela
def realizar_projecao():
    # Obter o nome do fundo selecionado
    fundo_name = fundo_combobox.get()  # Obtem o valor selecionado no combobox
    future_steps = future_steps_entry.get()  # Obtem o número de dias para projeção

    # Validação do fundo selecionado
    if not fundo_name:
        messagebox.showerror("Erro", "Por favor, selecione um fundo.")
        return

    # Validação do número de dias para projeção
    if not future_steps.isdigit():
        messagebox.showerror("Erro", "Por favor, insira um número válido de dias para a projeção.")
        return

    # Checar se o fundo está presente nos resultados
    if fundo_name not in results:
        messagebox.showerror("Erro", f"Resultados para o fundo '{fundo_name}' não encontrados.")
        return

    # Criar uma nova janela para exibir o gráfico
    try:
        projection_window = tk.Toplevel(root)
        projection_window.title(f"Projeção de {future_steps} dias para o Fundo: {fundo_name}")
        projection_window.geometry("800x600")
        
        # Gerar projeção
        result = results[fundo_name]
        scaler = result['scaler']
        best_model = result['best_model']
        n_steps = 30

        # Recuperar sequência final
        last_sequence = result['data']['Rentabilidade_Acumulada'].values[-n_steps:]
        last_sequence_scaled = scaler.transform(last_sequence.reshape(-1, 1)).flatten()

        # Criar novas datas para projeção
        last_date = result['data']['DT_COMPTC'].iloc[-1]
        future_dates = pd.date_range(start=last_date + pd.Timedelta(days=1), periods=int(future_steps))

        # Gerar projeções
        future_predictions = []
        for _ in range(int(future_steps)):
            input_sequence = np.array(last_sequence_scaled[-n_steps:]).reshape(1, n_steps, 1)
            next_pred_scaled = best_model.predict(input_sequence)[0, 0]
            next_pred = scaler.inverse_transform([[next_pred_scaled]])[0, 0]
            future_predictions.append(next_pred)
            last_sequence_scaled = np.append(last_sequence_scaled, next_pred_scaled)

        # Gerar gráfico de projeção
        fig, ax = plt.subplots(figsize=(10, 5))
        ax.plot(result['data']['DT_COMPTC'], result['data']['Rentabilidade_Acumulada'], label='Real')
        ax.plot(future_dates, future_predictions, label='Projeção', linestyle='--')
        ax.set_title(f'Projeção para o Fundo: {fundo_name}')
        ax.set_xlabel('Data')
        ax.set_ylabel('Rentabilidade Acumulada')
        ax.legend()

        # Integrar o gráfico na nova janela
        canvas_fig = FigureCanvasTkAgg(fig, master=projection_window)
        canvas_fig.get_tk_widget().pack(expand=True, fill=tk.BOTH)
        canvas_fig.draw()

    except Exception as e:
        messagebox.showerror("Erro", f"Ocorreu um erro ao realizar a projeção: {e}")
    
# Função para preencher os fundos no combobox ao abrir a tela
def preencher_fundos_combobox():
    if 'Razão Social' in data.columns:
        fundos_disponiveis = sorted(data['Razão Social'].unique())
        fundo_combobox['values'] = fundos_disponiveis
    else:
        messagebox.showerror("Erro", "Coluna 'Razão Social' não encontrada no DataFrame.")
        

 # Função para abrir a nova tela de previsões e projeções
def abrir_tela_previsoes():
    # Criação da nova janela
    tela_previsoes = tk.Toplevel(root)
    tela_previsoes.title("Modelos e Projeções")
    tela_previsoes.geometry("800x600")

    # Frame para previsões
    frame_previsoes = tk.Frame(tela_previsoes, pady=20)
    frame_previsoes.pack(fill=tk.X)

    tk.Label(frame_previsoes, text="Gráficos de Modelos", font=("Arial", 14)).pack(pady=10)

    btn_previsoes = ttk.Button(frame_previsoes, text="Gerar Modelos para Todos os Fundos", command=previsoes)
    btn_previsoes.pack(pady=10)

    # Separador
    ttk.Separator(tela_previsoes, orient="horizontal").pack(fill=tk.X, pady=20)

    # Frame para projeções
    frame_projecoes = tk.Frame(tela_previsoes, pady=20)
    frame_projecoes.pack(fill=tk.X)

    tk.Label(frame_projecoes, text="Projeções para Fundos Específicos", font=("Arial", 14)).pack(pady=10)

    tk.Label(frame_projecoes, text="Selecione o Fundo:").pack()

    # Preencher o combobox com os fundos a partir do DataFrame
    fundos_disponiveis = sorted(data['Razão Social'].unique())
    global fundo_combobox
    fundo_combobox = ttk.Combobox(frame_projecoes, values=fundos_disponiveis, state="readonly")
    fundo_combobox.pack(pady=5)

    tk.Label(frame_projecoes, text="Dias para Projeção:").pack()
    global future_steps_entry
    future_steps_entry = tk.Entry(frame_projecoes)
    future_steps_entry.pack(pady=5)

    btn_projecao = ttk.Button(frame_projecoes, text="Realizar Projeção", command=realizar_projecao)
    btn_projecao.pack(pady=10)

# Criando a janela principal com Tkinter
root = tk.Tk()
root.title("Seleção de Categoria")

# Botão para abrir a janela de categorias
btn_categoria = ttk.Button(root, text="Selecionar Categoria", command=selecionar_categoria)
btn_categoria.pack(pady=10)


# # Botão para gerar o gráfico de rentabilidade acumulada
# btn_rentabilidade = ttk.Button(root, text="Gerar Gráfico de Rentabilidade Acumulada", command=rentabilidade_acumulada)
# btn_rentabilidade.pack(pady=10)

root.mainloop()
```
